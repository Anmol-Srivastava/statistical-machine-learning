```python
import numpy as np
import pandas as pd 
```

```python
# load data 
df = pd.read_csv('../Resources/Auto.csv')

# remove missing values
df = df.dropna()
df = df[df.horsepower != '?']
```

```python
# problem 5, part a
cols = list(df.columns)
```

```python
# Problem 5, Part (b)

# name is the only qualitative predictor 
quants = cols.copy()
quants.remove('name')
data = [] 

# show range values
for col in quants: 
    values = pd.to_numeric(df[col])
    rng = max(values) - min(values)
    data.append([col, rng])

pd.DataFrame(data, columns = ['Predictor', 'Range'])
```

```python
# Problem 5, Part (c)

data = [] 

# show mean and SD values 
for col in quants: 
    values = pd.to_numeric(df[col])
    mean = np.mean(values)
    sd = np.std(values)
    data.append([col, mean, sd])

pd.DataFrame(data, columns = ['Predictor', 'Mean', 'St. Dev.'])
```

```python
# Problem 5, Part (d)

# drop last 50 obs
copy = df.copy()
copy = copy.head(-50)
data = [] 

# show new statistics in tabular format 
for col in quants:
    values = pd.to_numeric(copy[col])
    mean = np.mean(values)
    sd = np.std(values)
    rng = max(values) - min(values)
    data.append([col, rng, mean, sd])

pd.DataFrame(data, columns = ['Predictor', 'Range', 'Mean', 'St. Dev.'])
```
